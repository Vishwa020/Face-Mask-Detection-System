# Face mask detection in live environment

```python
# Importing libraries.
from cv2 import cv2
import numpy as np
from keras.models import load_model
from keras.preprocessing import image

# Loading the trained model.
model = load_model(r"E:\Projects\Face and face mask recognition model\model_final.h5")

# Creating dictionary of the labels with the rectangle color.
labels_dict = {0: 'with_mask', 1: 'without_mask'}
color_dict = {0: (0, 255, 0), 1: (255, 0, 0)}  # BGR format

size = 4  # Rectangle size
cap = cv2.VideoCapture(0)  # Using webcam 0

# Loading the haarcascade xml file.
classifier = cv2.CascadeClassifier(r"E:\Projects\Face and face mask recognition model\haarcascade_frontalface_default.xml")

# Detecting face mask until enter key is pressed.
# detecting while true (always).

while cap.isOpened():
    _, img = cap.read()

    img = cv2.flip(img, 1, 1)
    face = classifier.detectMultiScale(img, scaleFactor=1.1, minNeighbors=4)  # reduce the false positives
    
    for(x, y, w, h) in face:
        face_img = img[y:y+h, x:x+w]
        cv2.imwrite(r'E:\Projects\Face and face mask recognition model\temp.jpg', face_img)

        test_image = image.load_img(r'E:\Projects\Face and face mask recognition model\temp.jpg', target_size=(150, 150, 3))
        test_image = image.img_to_array(test_image)
        test_image = np.expand_dims(test_image, axis=0)
        
        label = model.predict_classes(test_image)[0][0] 
        print("label ", label)

        cv2.rectangle(img, (x, y), (x+w, y+h), color_dict[label], 2)
        cv2.rectangle(img, (x, y-40), (x+w, y), color_dict[label], -1)
        cv2.putText(img, labels_dict[label], (x, y-10), cv2.FONT_HERSHEY_SIMPLEX, 0.8, (255, 255, 255), 2)

    # Shows the live video stream.
    cv2.imshow('LIVE Capture', img)
    key = cv2.waitKey(10) # waits 10 miliseconds
    
    # If Enter key is press then break out of the loop.
    if key == 13:  # The Enter key
        break

# Stop video.
# Releases the webcam.
cap.release()

# Close all started windows.
cv2.destroyAllWindows()
```
